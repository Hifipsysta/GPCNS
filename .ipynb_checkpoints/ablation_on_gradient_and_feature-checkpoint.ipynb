```python
def compute_BWT(acc_str, size):
    import re
    pattern = r'\d+\.\d{1}'
    matches = re.findall(pattern, acc_str)
    acc_mat = np.array(matches, dtype=np.float32).reshape(size,size)
    bwt_lst = []
    for i in np.arange(size-1,0,-1):
        bwt=np.mean((acc_mat[-i]-np.diag(acc_mat))[:-i])
        bwt_lst.append(bwt.round(3))
    return bwt_lst
```
