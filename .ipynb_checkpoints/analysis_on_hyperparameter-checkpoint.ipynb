```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha_list = [1, 1.5, 3, 4.5, 6]
SEGPCNS_mini_acc = [66.07, 66.06, 65.60, 65.83, 64.90]
SEGPCNS_mini_bwt = [0.32, 0.13, -0.02, 0.23,-0.26]
scale_data = [i*1.5 for i in range(len(alpha_list))]
```

```python
delta_acc_segpcns = 62
scale_bwt_segpcns = 5
```

```python
fig, ax = plt.subplots()   
plt.xticks([]) 
bars_acc = ax.bar(scale_data,np.array(SEGPCNS_mini_acc)-delta_acc_segpcns,lw=0.5,fc="darkcyan",label="Phone")
bars_bwt = plt.bar(scale_data,np.array(SEGPCNS_mini_bwt)*scale_bwt_segpcns,lw=0.5,fc="orange",label="Android") #bottom=bwt_data
plt.ylim(-4, 6)

for bar in bars_acc:  
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval+0.2,  
            round(yval, 2)+delta_acc_segpcns,               
            va='bottom',                 
            ha='center',                 
            fontsize=14, color='black')  
    
for bar in bars_bwt:  
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval-0.7,  
            round(yval/scale_bwt_segpcns, 2),              
            va='bottom',                 
            ha='center',                 
            fontsize=14, color='black')  
    
ax.set_xticks([bar.get_x() + bar.get_width() / 2 for bar in bars_acc], 
              [r'$\alpha={}$'.format(alpha_list[0]), 
               r'$\alpha={}$'.format(alpha_list[1]),
               r'$\alpha={}$'.format(alpha_list[2]),
               r'$\alpha={}$'.format(alpha_list[3]),
               r'$\alpha={}$'.format(alpha_list[4])])
ax.set_yticks([-4,-2, 4,6],[-4/scale_bwt_segpcns,-2/scale_bwt_segpcns, 4+delta_acc_segpcns,6+delta_acc_segpcns])

ax.axhline(y=0.0, c='black', ls='-', lw=1)


ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')   
# ax.xaxis.set_ticks_position('bottom') 
# ax.yaxis.set_ticks_position('left') 


plt.yticks(fontproperties='Times New Roman', size=20)
plt.xticks(fontproperties='Times New Roman', size=18)
plt.ylabel('BWT(%)          ACC(%)', size=20)
plt.show()
```

```python
alpha_list = [1, 1.5, 3, 4.5, 6]
GPCNS_mini_acc = [61.63, 63.06, 63.78, 60.52, 60.50]
GPCNS_mini_bwt = [-4.74, -3.03, -1.97, -5.05,-6.53]
scale_data = [i*1.5 for i in range(len(alpha_list))]
```

```python
scale_bwt_gpcns = 2
delta_acc_gpcns = 60
```

```python
fig, ax = plt.subplots()   
plt.xticks([]) 
bars_acc = ax.bar(scale_data,np.array(GPCNS_mini_acc)-delta_acc_gpcns,lw=0.5,fc="darkcyan",label="Phone")
bars_bwt = plt.bar(scale_data,np.array(GPCNS_mini_bwt)/scale_bwt_gpcns,lw=0.5,fc="orange",label="Android") #bottom=bwt_data
plt.ylim(-5, 6)

for bar in bars_acc:  
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval+0.2,  
            round(yval, 2)+delta_acc_gpcns, 
            va='bottom',                 
            ha='center',                 
            fontsize=14, color='black')  
    
for bar in bars_bwt:  
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval-0.7,  
            round(yval*scale_bwt_gpcns, 2),              
            va='bottom',                 
            ha='center',                 
            fontsize=14, color='black')  
    
ax.set_xticks([bar.get_x() + bar.get_width() / 2 for bar in bars_acc], 
              [r'$\alpha={}$'.format(alpha_list[0]), 
               r'$\alpha={}$'.format(alpha_list[1]),
               r'$\alpha={}$'.format(alpha_list[2]),
               r'$\alpha={}$'.format(alpha_list[3]),
               r'$\alpha={}$'.format(alpha_list[4])])
ax.set_yticks([-5,-2, 4,6],[-5*scale_bwt_gpcns,-2*scale_bwt_gpcns, delta_acc_gpcns+4,delta_acc_gpcns+6])

ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')  
ax.axhline(y=0.0, c='black', ls='-', lw=1)

plt.yticks(fontproperties='Times New Roman', size=20)
plt.xticks(fontproperties='Times New Roman', size=18)
plt.ylabel('BWT (%)        ACC (%)', size=22)
plt.show()
```
