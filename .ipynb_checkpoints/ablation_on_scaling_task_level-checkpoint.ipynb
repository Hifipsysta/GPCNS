```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
taskid_10 = [i for i in np.arange(1,11)]
select_id = [1,4,7,10]
```

```python
plt.yticks(fontproperties='Times New Roman', size=18)
plt.xticks(fontproperties='Times New Roman', size=18)

plt.plot(taskid_10, [76.7, 72.1, 75.0, 73.6, 78.3, 76.2, 76.9, 74.9, 78.3, 80.5], label='SE-GPCNS', linestyle='--', color='red', marker='v')
plt.plot(taskid_10, [70.3, 66.3, 74.0, 72.4, 75.3, 77.1, 76.1, 74.5, 76.4, 81.6], label='GPCNS', linestyle='--', color='blue', marker='o')
plt.plot(taskid_10, [77.0, 71.9, 75.1, 70.2, 76.1, 71.4, 73.2, 72.3, 72.0, 77.2], label='SE-GPCNS (w/o scale)', color='orange',marker='v')
plt.plot(taskid_10, [68.3, 65.6, 73.0, 70.4, 74.5, 75.3, 75.9, 74.3, 76.7, 80.6], label='GPCNS (w/o scale)', color='green', marker='o')
plt.ylim(30, 100)
plt.xticks(select_id, [str(i) for i in select_id])
plt.xlabel('Task ID', fontsize=18)
plt.ylabel('ACC (%)', fontsize=18)
plt.legend(loc=3, fontsize=11.5)
```
